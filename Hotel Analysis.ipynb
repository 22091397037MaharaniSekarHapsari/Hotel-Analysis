```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load dataset
df = pd.read_csv('/content/midterm_hotel_data_cleansing.csv')


# Menghitung persentase pembatalan
cancel_percentages = df['is_canceled'].value_counts(normalize=True) * 100


# Menampilkan hasil persentase
print("Persentase Tidak Dibatalkan:", cancel_percentages[0], "%")
print("Persentase Dibatalkan:", cancel_percentages[1], "%")


# Membuat grafik batang distribusi dengan persentase
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='is_canceled', palette='Blues')
plt.title('Distribusi Pembatalan Reservasi')
plt.xlabel('Status Pembatalan')
plt.ylabel('Jumlah')
plt.xticks([0, 1], ['Tidak Dibatalkan', 'Dibatalkan'])


# Menambahkan teks persentase pada setiap batang
for index, value in enumerate(cancel_percentages):
    plt.text(index, df['is_canceled'].value_counts()[index], f'{value:.1f}%', ha='center', va='bottom', fontsize=12)


plt.show()
```

```python
#Module 1

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# 1. Membuat kolom baru untuk mengelompokkan lead_time
bins = [0, 7, 14, 21, 30, 40, 50, 60, 70, 80, 90, 180, 365, df['lead_time'].max()]
labels = ['0-7', '8-14', '15-21', '22-30','31-40', '41-50','51-60', '61-70', '71-80','81-90', '91-180', '181-365', '365+']
df['lead_time_group'] = pd.cut(df['lead_time'], bins=bins, labels=labels, right=False)

# 2. Menghitung rata-rata pembatalan berdasarkan kelompok lead_time
cancellation_rate_by_group = df.groupby('lead_time_group')['is_canceled'].mean() * 100

# 3. Visualisasi hubungan lead_time dan is_canceled berdasarkan kelompok
plt.figure(figsize=(8, 6))
sns.lineplot(x=cancellation_rate_by_group.index, y=cancellation_rate_by_group.values, marker='o', color='purple')
plt.title('Rata-rata Tingkat Pembatalan Berdasarkan Kelompok Lead Time')
plt.xlabel('Kelompok Lead Time (hari)')
plt.ylabel('Persentase Pembatalan (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd


# Mengisi nilai kosong dengan 0
df['children'] = df['children'].fillna(0)
df['adults'] = df['adults'].fillna(0)
df['total_of_special_requests'] = df['total_of_special_requests'].fillna(0)
df['stays_in_weekend_nights'] = df['stays_in_weekend_nights'].fillna(0)
df['stays_in_week_nights'] = df['stays_in_week_nights'].fillna(0)

# Mengisi nilai kosong di kolom lead_time dan adr dengan median
df['lead_time'] = df['lead_time'].fillna(df['lead_time'].median())
df['adr'] = df['adr'].fillna(df['adr'].median())

# Mengisi nilai kosong di kolom agent dan company dengan modus
df['agent'] = df['agent'].fillna('Tidak Ada')
df['company'] = df['company'].fillna('Tidak Ada')

# Mengisi nilai kosong di kolom country dengan 'Tidak Diketahui'
df['country'] = df['country'].fillna('Tidak Diketahui')
```

```python
# mengubah tipe data date
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

# mengubah tipe data children, adults, total_of_special_requests
df['children'] = df['children'].astype(int)
df['adults'] = df['adults'].astype(int)
df['total_of_special_requests'] = df['total_of_special_requests'].astype(int)
```

```python
import pandas as pd

# Hapus baris di mana 'adults' dan 'children' keduanya bernilai 0
df = df[~((df['adults'] == 0) & (df['children'] == 0))]

# Hapus baris di mana 'stay_week' dan 'stay_weekend' keduanya bernilai 0
df = df[~((df['stays_in_week_nights'] == 0) & (df['stays_in_week_nights'] == 0))]
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Memisahkan data pembatalan untuk City Hotel dan Resort Hotel
canceled_reservations_city = df[(df['is_canceled'] == 1) & (df['hotel'] == 'City Hotel')]
canceled_reservations_resort = df[(df['is_canceled'] == 1) & (df['hotel'] == 'Resort Hotel')]

# Mengelompokkan pembatalan berdasarkan negara untuk masing-masing jenis hotel
canceled_by_country_city = canceled_reservations_city.groupby('country').size().reset_index(name='cancellations')
canceled_by_country_resort = canceled_reservations_resort.groupby('country').size().reset_index(name='cancellations')

# Mengurutkan hasil dan ambil 10 negara teratas untuk masing-masing jenis hotel
top_10_countries_city = canceled_by_country_city.sort_values(by='cancellations', ascending=False).head(10)
top_10_countries_resort = canceled_by_country_resort.sort_values(by='cancellations', ascending=False).head(10)

# Menambahkan kolom 'hotel' untuk membedakan jenis hotel
top_10_countries_city['hotel'] = 'City Hotel'
top_10_countries_resort['hotel'] = 'Resort Hotel'

# Menggabungkan kedua data menjadi satu dataframe dalam format panjang
top_10_combined = pd.concat([top_10_countries_city, top_10_countries_resort])

# Menampilkan data gabungan
print("Data Pembatalan City Hotel dan Resort Hotel (Gabungan):")
print(top_10_combined)

# Visualisasi dengan Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_combined, x='country', y='cancellations', hue='hotel', palette=['skyblue', 'salmon'])

# Menyesuaikan tampilan
plt.xticks(rotation=45)
plt.title('10 Negara dengan Jumlah Pembatalan Tertinggi - City Hotel dan Resort Hotel')
plt.xlabel('Negara')
plt.ylabel('Jumlah Pembatalan')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from scipy.stats import shapiro, levene, mannwhitneyu

city_hotel_adr = df[df['hotel'] == 'City Hotel']['adr']
resort_hotel_adr = df[df['hotel'] == 'Resort Hotel']['adr']

#uji normalitas
normality_city = shapiro(city_hotel_adr)
normality_resort = shapiro(resort_hotel_adr)
print(f"Normality City Hotel ADR: {normality_city}")
print(f"Normality Resort Hotel ADR: {normality_resort}")

#uji kesamaan varians
levene_test = levene(city_hotel_adr, resort_hotel_adr)
print(f"Levene's Test for Equal Variances: {levene_test}")

#uji A/B testing dengan Mann-Whitney U Test
mannwhitney_test = mannwhitneyu(city_hotel_adr, resort_hotel_adr)
print(f"Mann-Whitney U Test result: {mannwhitney_test}")

# Tingkat signifikansi (5%)
alpha = 0.05

# Kesimpulan Uji
p_value = mannwhitney_test.pvalue
if p_value < alpha:
    print("Hipotesis nol (H0) ditolak.")
    print("Terdapat perbedaan signifikan antara rata-rata ADR City Hotel dan Resort Hotel.")
else:
    print("Hipotesis nol (H0) diterima.")
    print("Tidak terdapat perbedaan signifikan antara rata-rata ADR City Hotel dan Resort Hotel.")
```
